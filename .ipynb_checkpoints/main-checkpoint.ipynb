# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder,StandardScaler,FunctionTransformer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('test.csv')
```

# Data Preprocessing

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns=['id','homepage','imdb_id','original_language','original_title','poster_path','tagline','overview','belongs_to_collection'],inplace=True)
```

```python
import ast
cols =  ['genres','production_companies','production_countries','spoken_languages','Keywords','cast','crew']
for col in cols:
    df[col] = df[col].apply(lambda x:ast.literal_eval(x)if isinstance(x,str)else x)
```

```python
df['popularity'] = df['popularity'].astype(int)
df['release_date'] = pd.to_datetime(df['release_date'],errors='coerce')
```

```python
df['runtime'] = pd.to_numeric(df['runtime'],errors='coerce')
df['budget'] = pd.to_numeric(df['budget'],errors='coerce')
```

```python
df.head()
```

```python
def extract_names(x):
    if isinstance(x,list):
        return [i['name'] for i in x if isinstance(i,dict)]
```

```python
for col in cols:
    df[col] = df[col].apply(extract_names)
```

```python
df
```

```python
obj_col = df.select_dtypes('object').astype('str')
for col in obj_col:
    df[col] = df[col].apply(lambda x:','.join(x) if isinstance(x,list)else x)
```

```python
df
```

# Handling Null Values

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.isnull())
plt.show()
```

```python
for col in df:
    if df[col].isnull().sum()>5:
        df[col] = df[col].fillna('not given')
df['release_date'] = df['release_date'].fillna(df['release_date'].mode)
df['runtime'] = df['runtime'].fillna(df['runtime'].mode)
df['status'] = df['status'].fillna(df['status'].mode)
df['title'] = df['title'].fillna('not given')
```

```python
df.isnull().sum()
```
