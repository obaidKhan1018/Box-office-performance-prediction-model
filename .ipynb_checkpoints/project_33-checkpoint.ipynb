```python
#checking missing values
# Check for missing values
missing_values = data.isnull().sum()

# Display columns with missing values
print(missing_values[missing_values > 0])
```

```python
# Step 1: Identify values in budget starting with '0.' and treat them as missing
# Convert those values to NaN
data.loc[data['budget'].astype(str).str.startswith('0.'), 'budget'] = pd.NA
data
```

```python
# List of numerical columns
numerical_columns = ['budget', 'popularity', 'runtime']

# 1. Impute missing values with mean
num_imputer = SimpleImputer(strategy='mean')
data[numerical_columns] = num_imputer.fit_transform(data[numerical_columns])

# Display the count of missing values for numerical columns to verify
print(data[numerical_columns].isnull().sum())
```

```python
# List of categorical columns
categorical_columns = [
    'belongs_to_collection', 'genres', 'production_companies',
    'production_countries', 'original_language', 'spoken_languages', 'status'
]

# Create an imputer for categorical data, using the most frequent strategy
cat_imputer = SimpleImputer(strategy='most_frequent')

# Apply imputer to categorical columns and store back in the original DataFrame
data[categorical_columns] = cat_imputer.fit_transform(data[categorical_columns])

# Check if missing values are handled
print(data[categorical_columns].isnull().sum())
```

```python
# List of columns to drop
columns_to_drop = ['id', 'imdb_id', 'poster_path', 'homepage', 'tagline', 'overview']

# Drop the columns from the DataFrame
data.drop(columns=columns_to_drop, inplace=True)

# Display the first few rows to verify the columns are dropped
print(data.head())
```

```python
#verify all misising values has been handled
# Check for any remaining missing values in the entire dataset
missing_values = data.isnull().sum()

# Display columns with missing values, if any
print("Missing values in each column:\n", missing_values[missing_values > 0])
```

```python
# Columns with missing values
columns_with_missing_values = ['release_date', 'title', 'Keywords', 'cast', 'crew']

# 1. Impute 'release_date' and 'title' with the most frequent value (mode)
cat_imputer = SimpleImputer(strategy='most_frequent')
data[['release_date', 'title']] = cat_imputer.fit_transform(data[['release_date', 'title']])

# 2. Impute 'Keywords', 'cast', and 'crew' with the most frequent value (mode)
data[['Keywords', 'cast', 'crew']] = cat_imputer.fit_transform(data[['Keywords', 'cast', 'crew']])

# Verify that missing values are handled
print(data[columns_with_missing_values].isnull().sum())
```

```python
#verify all misising values has been handled
# Check for any remaining missing values in the entire dataset
missing_values = data.isnull().sum()

# Display columns with missing values, if any
print("Missing values in each column:\n", missing_values[missing_values > 0])
```

```python
#Label Encoding for the status Column
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Encode the 'Sex' column
data['status'] = le.fit_transform(data['status'])

# Verify the encoding
print(data[['status']].head(10))
```

```python
#apply one hot coding on genres

# Sample data (assuming 'data' is your DataFrame)
data = pd.DataFrame({
    'genres': ['Action|Adventure', 'Comedy|Drama', 'Action', 'Comedy']
})

# Apply one-hot encoding to 'genres'
genres_encoded = data['genres'].str.get_dummies(sep='|')

# Join the encoded columns to the original DataFrame
data = data.join(genres_encoded)

# Drop the original 'genres' column
data = data.drop(columns=['genres'])

# Display the first few rows of the modified DataFrame
print(data.head())
```

```python
#one hot coding on belongd to collection

# Sample data (assuming 'data' is your DataFrame)
data = pd.DataFrame({
    'belongs_to_collection': ['Collection1', 'Collection2', 'Collection1', None]
})

# Apply one-hot encoding to 'belongs_to_collection'
belongs_to_collection_encoded = pd.get_dummies(data['belongs_to_collection'], prefix='collection')

# Join the encoded columns to the original DataFrame
data = data.join(belongs_to_collection_encoded)

# Drop the original 'belongs_to_collection' column
data = data.drop(columns=['belongs_to_collection'])

# Display the first few rows of the modified DataFrame
print(data.head())
```

```python
# one hot coding on production comapnies

data = pd.DataFrame({
    'production_companies': ['Company1|Company2', 'Company3', 'Company1', 'Company4|Company5']
})

# Apply one-hot encoding to 'production_companies'
production_companies_encoded = data['production_companies'].str.get_dummies(sep='|')

# Join the encoded columns to the original DataFrame
data = data.join(production_companies_encoded)

# Drop the original 'production_companies' column
data = data.drop(columns=['production_companies'])

# Display the first few rows of the modified DataFrame
print(data.head())
```
